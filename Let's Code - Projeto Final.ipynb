```python
def get_dataset(y, labels):
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label': labels[i],
                'data': y[i]
            })
        return datasets
    else:
        return[
            {
                'label': labels[0],
                'data': y
            }
        ]
```

```python
def set_title(title=''):
    if title != '':
        display = 'true'
    else:
        display = 'false'
        
    return {
        'title': title,
        'display': display
    }
```

```python
def create_chart(x, y, labels, kind='bar', title=''):
    datasets = get_dataset(y, labels)
    options = set_title(title)
    
    chart = {
        'type': kind,
        'data': {
            'labels': x,
            'datasets': datasets
        },
        'options': options
    }
    
    return chart
```

```python
def get_api_chart(chart):
    base_url = 'https://quickchart.io/chart'
    response = r.get(f'{base_url}/?c={str(chart)}')
    
    return response.content
```

```python
def save_image(path, content): 
    with open(path, 'wb') as image:
        image.write(content)
```

```python
from PIL import Image
from IPython.display import display
```

```python
def display_image(path):
    image_pil = Image.open(path)
    display(image_pil)
```

```python
y_data_one = []
for obs in final_data[1::10]:
    y_data_one.append(obs[CONFIRMED])

y_data_two = []
for obs in final_data[1::10]:
    y_data_two.append(obs[RECOVERED])
    
labels = ['Confirmed', 'Recovered']

x = []
for obs in final_data[1::10]:
    x.append(obs[DATE].strftime('%d/%m/%Y'))
    
chart = create_chart(x, [y_data_one, y_data_two], labels, title = 'Graph Confirmed x Recovered')
chart_content = get_api_chart(chart)
save_image('my_first_graph.png', chart_content)
display_image('my_first_graph.png')

```

```python
from urllib.parse import quote
```

```python
def get_api_qrcode(link):
    text = quote(link) # parsing do link para url
    url_base = 'https://quickchart.io/qr'
    resp = r.get(f'{url_base}?text={text}')
    
    return resp.content
```

```python
base_url = 'https://quickchart.io/chart'
link = f'{base_url}/?c={str(chart)}'

save_image('qr-code.png', get_api_qrcode(link))
display_image('qr-code.png')
```
